### Removing **outliers**

```python
def outliers_detection_iqr(df):
    Q1 = df.quantile(0.25)
    Q3 = df.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df < lower_bound) | (df > upper_bound)]
    return outliers
```

```python
def remove_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    filtered_df = data[(data[column] >= lower_bound) & (data[column] <= upper_bound)]
    return filtered_df
```

```python
numeric_cols = [ 'Vibration_mms', 'Sound_dB', 'Oil_Level_pct', 'Coolant_Level_pct', 'Power_Consumption_kW','Temperature_C']

df_clean = df.copy()
for col in numeric_cols:
    df_clean = remove_outliers_iqr(df_clean, col)

print(f'Shape after removing outliers in multiple columns: {df_clean.shape}')
```

# Advanced Data Analysis: Time-Series & PCA

```python
import matplotlib as mpl
from statsmodels.tsa.seasonal import seasonal_decompose
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# prevent OverflowError in Agg backend
mpl.rcParams['agg.path.chunksize'] = 10000
mpl.rcParams['path.simplify'] = True
mpl.rcParams['path.simplify_threshold'] = 0.1 

# --- Time-Series Analysis ---
plt.figure(figsize=(12, 5))
plt.plot(
    df_clean["Operational_Hours"],
    df_clean["Temperature_C"],
    alpha=0.3,
    linewidth=0.8,    
    label="Temperature"
)
plt.title("Temperature vs Operational Hours (Sampled)")
plt.xlabel("Operational Hours")
plt.ylabel("Temperature (Â°C)")
plt.legend()
plt.tight_layout()
plt.show()


df_sorted = df_clean.sort_values("Operational_Hours").reset_index(drop=True)


result = seasonal_decompose(
    df_sorted["Temperature_C"].iloc[:2000], period=200, model="additive"
)
fig = result.plot()
fig.set_size_inches(12, 8)
plt.suptitle("Seasonal Decomposition of Temperature_C", fontsize=14, weight="bold")
plt.tight_layout()
plt.show()

# --- Anomaly Detection ---
mu, sigma = df_clean["Temperature_C"].mean(), df_clean["Temperature_C"].std()
df_clean["Temp_Anomaly"] = (
    (df_clean["Temperature_C"] > mu + 3 * sigma) |
    (df_clean["Temperature_C"] < mu - 3 * sigma)
)

plt.figure(figsize=(12, 5))
plt.plot(
    df_clean["Operational_Hours"],
    df_clean["Temperature_C"],
    alpha=0.4,
    linewidth=0.8,
    label="Temperature"
)
plt.scatter(
    df_clean.loc[df_clean["Temp_Anomaly"], "Operational_Hours"],
    df_clean.loc[df_clean["Temp_Anomaly"], "Temperature_C"],
    color="red", label="Anomaly"
)
plt.title("Temperature Anomaly Detection")
plt.xlabel("Operational Hours")
plt.ylabel("Temperature (Â°C)")
plt.legend()
plt.tight_layout()
plt.show()

```

```python

features = [
    "Operational_Hours", "Temperature_C", "Vibration_mms",
    "Sound_dB", "Oil_Level_pct", "Coolant_Level_pct",
    "Power_Consumption_kW", "Last_Maintenance_Days_Ago",
    "Maintenance_History_Count", "Failure_History_Count",
    "Error_Codes_Last_30_Days", "AI_Override_Events",
    "Remaining_Useful_Life_days"
]

X = df_clean[features]

# Standardize
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# PCA
pca = PCA(n_components=5)
X_pca = pca.fit_transform(X_scaled)

explained_var = pca.explained_variance_ratio_

# Plot Explained Variance
plt.figure(figsize=(8, 5))
sns.barplot(
    x=[f"PC{i+1}" for i in range(len(explained_var))],
    y=explained_var,
    palette="viridis"
)
plt.title("PCA - Explained Variance Ratio")
plt.ylabel("Variance Ratio")
plt.tight_layout()
plt.show()

# Show most important features contributing to each PC
loadings = pd.DataFrame(
    pca.components_.T,
    columns=[f"PC{i+1}" for i in range(5)],
    index=features
)

print("Top Contributing Features per Principal Component:")
display(loadings.abs().sort_values(by="PC1", ascending=False).head(10))

# Scatter Plot of first two PCs
plt.figure(figsize=(8, 6))
sns.scatterplot(
    x=X_pca[:, 0],
    y=X_pca[:, 1],
    hue=df_clean["Failure_Within_7_Days"],
    palette=['#2ecc71', '#e74c3c'],
    alpha=0.5
)
plt.title("PCA Projection (PC1 vs PC2)")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.legend(title="Failure within 7 Days")
plt.tight_layout()
plt.show()
```

# Feature engineering

```python
df_clean = df_clean.sort_values("Operational_Hours").reset_index(drop=True)

df_clean["Temp_RateChange"] = df_clean["Temperature_C"].diff() / df_clean["Operational_Hours"].diff()
df_clean["Vibration_RateChange"] = df_clean["Vibration_mms"].diff() / df_clean["Operational_Hours"].diff()
df_clean["Pressure_RateChange"] = df_clean["Power_Consumption_kW"].diff() / df_clean["Operational_Hours"].diff()

df_clean[["Temp_RateChange", "Vibration_RateChange", "Pressure_RateChange"]] = \
    df_clean[["Temp_RateChange", "Vibration_RateChange", "Pressure_RateChange"]].fillna(0)


df_clean["Temp_Degradation"] = df_clean["Temperature_C"] - df_clean["Temperature_C"].iloc[0]
df_clean["Vibration_Degradation"] = df_clean["Vibration_mms"] - df_clean["Vibration_mms"].iloc[0]

# Rolling slope (trend) of temperature and vibration (using linear regression on a rolling window)
window = 50  
df_clean["Temp_Trend"] = df_clean["Temperature_C"].rolling(window).apply(lambda x: pd.Series(x).diff().mean(), raw=False)
df_clean["Vibration_Trend"] = df_clean["Vibration_mms"].rolling(window).apply(lambda x: pd.Series(x).diff().mean(), raw=False)


# --- Rolling Statistics (Noise Reduction + Trend Capture)
df_clean["Temp_MA_50"] = df_clean["Temperature_C"].rolling(window=50).mean()
df_clean["Temp_STD_50"] = df_clean["Temperature_C"].rolling(window=50).std()

df_clean["Vibration_MA_50"] = df_clean["Vibration_mms"].rolling(window=50).mean()
df_clean["Vibration_STD_50"] = df_clean["Vibration_mms"].rolling(window=50).std()

df_clean["Pressure_MA_50"] = df_clean["Power_Consumption_kW"].rolling(window=50).mean()
df_clean["Pressure_STD_50"] = df_clean["Power_Consumption_kW"].rolling(window=50).std()

print("New engineered features created:")
print([
    "Temp_RateChange", "Vibration_RateChange", "Pressure_RateChange",
    "Temp_Degradation", "Vibration_Degradation",
    "Temp_Trend", "Vibration_Trend",
    "Temp_MA_50", "Temp_STD_50",
    "Vibration_MA_50", "Vibration_STD_50",
    "Pressure_MA_50", "Pressure_STD_50"
])

plt.figure(figsize=(12, 5))
plt.plot(df_clean["Operational_Hours"], df_clean["Temperature_C"], alpha=0.4, label="Temperature")
plt.plot(df_clean["Operational_Hours"], df_clean["Temp_MA_50"], color="red", label="Temp Moving Avg (50)")
plt.title("Temperature with Rolling Average")
plt.xlabel("Operational Hours")
plt.ylabel("Temperature (Â°C)")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
from scipy.fftpack import fft

# Focus only on actual sensor signals
sensor_cols = [
    "Temperature_C",
    "Vibration_mms",
    "Sound_dB",
    "Oil_Level_pct",
    "Coolant_Level_pct",
    "Power_Consumption_kW"
]

window_size = 10

for col in sensor_cols:
    if col in df.columns:
        df[f"{col}_min"] = df[col].rolling(window_size, min_periods=1).min()
        df[f"{col}_max"] = df[col].rolling(window_size, min_periods=1).max()


def compute_fft_features(signal):
    signal = signal.dropna().values
    if len(signal) == 0:
        return pd.Series([0, 0, 0], index=["fft_mean", "fft_peak", "fft_energy"])

    fft_vals = np.abs(fft(signal))
    fft_mean = np.mean(fft_vals)
    fft_peak = np.max(fft_vals)
    fft_energy = np.sum(fft_vals**2)

    return pd.Series([fft_mean, fft_peak, fft_energy],
                     index=["fft_mean", "fft_peak", "fft_energy"])


df_fft = compute_fft_features(df["Vibration_mms"])
print("FFT features (Vibration):")
print(df_fft)
```
